###### Assuming here that first row of dataset is earliest part of travel - this is currently the case

```python
trips_start = app.groupby('Trip ID').first()
```

```python
def get_hour(date):
    return date.hour

def get_day_of_week(date):
    return date.weekday()

def get_weekend(day_num):
    if day_num in [5,6]:
        return 1
    return 0

trips_start['Date'] = pd.to_datetime(trips_start.Date)

trips_start['hour_of_day'] = trips_start.apply(lambda row: get_hour(row['Date']), axis=1)
trips_start['day_of_week'] = trips_start.apply(lambda row: get_day_of_week(row['Date']), axis=1)
trips_start['weekend'] = trips_start.apply(lambda row: get_weekend(row['day_of_week']), axis=1)
```

```python
trips_end = app.groupby('Trip ID').last()
trip_start_end = trips_start.join(trips_end, lsuffix=' Start', rsuffix=' End')
trip_start_end['lat_delta'] = trip_start_end['Latitude End'] - trip_start_end['Latitude Start']
trip_start_end['lon_delta'] = trip_start_end['Longitude End'] - trip_start_end['Longitude Start']
trip_start_end['alt_delta'] = trip_start_end['Altitude End'] - trip_start_end['Altitude Start']
```

```python
trip_max = app.groupby('Trip ID').max()
```

```python
app_trip_info = trip_start_end.join(trip_max['Speed'])
```

```python
def cut_off_speed(speed):
    if speed>20:
        return 20
    return speed

app_trip_info['Speed'] = app_trip_info.apply(lambda row: cut_off_speed(row['Speed']), axis=1)
app_trip_info['Speed']
```

```python
def remove_time(date):
    return date.date()

app_trip_info['Date Day'] = app_trip_info.apply(lambda row: remove_time(row['Date Start']), axis=1)
```

```python
def get_raining(rain):
    if rain > 0:
        return 1
    return 0

weather['Raining'] = weather.apply(lambda row: get_raining(row['Precipitation_MM']), axis=1)
```

```python
weather['Date'] = pd.to_datetime(weather['Date'], format='%m-%d-%Y')
weather['Date'] = weather.apply(lambda row: remove_time(row['Date']), axis=1)
```

```python
app_final_features = pd.concat([
                                app_final_features,
                                pd.get_dummies(app_final_features.hour_of_day, prefix='hour'),
                                pd.get_dummies(app_final_features.day_of_week, prefix='day')
                               ],
                                axis=1
                               )
```

```python
columns = app_final_features.columns
columns_to_drop = ['hour_of_day', 'day_of_week']
columns_to_keep = [i for i in columns if i not in columns_to_drop]
app_final_features = app_final_features[columns_to_keep]
```
