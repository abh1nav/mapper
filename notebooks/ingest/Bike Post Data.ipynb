```python
import requests

f = open('/Users/asdf/workspaces/abh1nav/mapper/data/bike-posts/addresses.csv', 'r')

addresses = [line for line in f]
f.close()
```

```python
addresses = [l.replace('\n', '') for l in addresses]
```

```python
addresses
```

```python
def geocode(addr):
    url = 'https://maps.googleapis.com/maps/api/geocode/json?address=%s'
    r = requests.get(url % addr)
    return r.json()
```

```python
geocodes = {}
for a in addresses:
    result = geocode('%s, Toronto' % a)
    geocodes[a] = result
```

```python
from pymongo import MongoClient
mongo_conn = MongoClient(['mongo1.x.crowdriff.com'])
mongo_db = mongo_conn['geocodes']['bike_posts']
for g in geocodes.keys():
    gcode = geocodes[g]
    gcode['address'] = g
    mongo_db.insert(gcode)
```

```python
from pymongo import MongoClient
def clean_bike_post_data():
    conn = MongoClient(['mongo1.x.crowdriff.com'])
    src = conn['geocodes']['bike_posts_raw']
    out = conn['geocodes']['bike_posts']
    for doc in src.find(timeout=False):
        loc = doc['results'][0]['geometry']['location']
        out_doc = {
            'loc': {
                'type': 'Point',
                'coordinates': [ loc['lng'], loc['lat'] ]
            },
            'addr': doc['address']
        }
        out.save(out_doc)

clean_bike_post_data()
```
