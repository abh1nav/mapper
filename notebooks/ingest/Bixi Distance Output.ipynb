```python
from pymongo import MongoClient
conn = MongoClient()
db = conn['geocodes']
coll = db['bixi_distance_extract']
pts = db['bixi_extract']

f = open('/Users/asdf/workspaces/abh1nav/mapper/data/bixi-distances/output.csv', 'w')
# CSV output template: start, end, distance (km), start_lat, start_lng, start_elevation, end_lat, end_lng, end_elevation
tmpl = '%s, %s, %s, %s, %s, %s, %s, %s, %s\n'
for doc in coll.find():
    start = doc['_id']['start']
    end = doc['_id']['end']
    distance = doc['distance']
    start_pt = pts.find_one({'_id': start})
    end_pt = pts.find_one({'_id': end})
    
    if start_pt and end_pt and 'elevation' in start_pt and 'elevation' in end_pt:
        line = tmpl % (start, end, distance, start_pt['lat'], start_pt['lng'], start_pt['elevation'], end_pt['lat'], end_pt['lng'], end_pt['elevation'])
        print line
        f.write(line)
f.close()
```
